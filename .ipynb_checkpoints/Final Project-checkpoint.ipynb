## Preprocessing

**Imputation for numeric Outliers**

The code snippet below was found on this website and edited for the project's purposes.

https://careerfoundry.com/en/blog/data-analytics/how-to-find-outliers/#:~:text=Finding%20outliers%20using%20statistical%20methods,-Since%20the%20data&text=Using%20the%20IQR%2C%20the%20outlier,Q1%20(Q3%E2%80%93Q1).

```python
# replacing the -1 to 0
new_df = bank
new_df['pdays'] = new_df['pdays'].replace(-1, 0)
```

```python
#num_vars = ['age', 'balance','day','duration','campaign','pdays','previous']
# changing the outliers to averages of the data set
# this function essentially finds the IQR and anything below 25th percentile - 1.5 *IQR and 
# anything above the 75th percentile + 1.5*IQR will count as outliers

def impute_outliers_IQR(df, var):
    new_df = df
    q1=new_df[var].quantile(0.25)
    q3=new_df[var].quantile(0.75)
    IQR=q3-q1
    upper = q3+1.5*IQR
    lower = q1-1.5*IQR
    new_df[var] = np.where(new_df[var] > upper, new_df[var].mean(), np.where(new_df[var] < lower,new_df[var].mean(),new_df[var]))
    return new_df
```

```python
new_df = bank
new_df = impute_outliers_IQR(new_df, num_vars)
```

### Splitting the dataset

```python
# Split data 70/30
X = pd.DataFrame(new_df.loc[:, new_df.columns != 'y'])
y = new_df[['y']]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, train_size=0.875, random_state=1)
```

### Data transformation

```python
# checking the distribution of the training data, validation, and testing data
X_train[num_vars].hist(figsize=(12,12),bins=20)
plt.show()
```

```python
# using the yeojohnson transformation because it can take both negative and postive numbers
X_train[num_vars] = X_train[num_vars].apply(lambda x: yeojohnson(x)[0] , axis=0)
X_test[num_vars] = X_test[num_vars].apply(lambda x: yeojohnson(x)[0] , axis=0)
X_val[num_vars] = X_val[num_vars].apply(lambda x: yeojohnson(x)[0] , axis=0)
```

```python
# checking the distribution of the training data, validation, and testing data after transformations
X_train[num_vars].hist(figsize=(12,12),bins=20)
plt.show()
```

### Scaling and One Hot Encoding

Creating the pipelines to scale the numeric data and one hot encode the categorical ones.

```python
# pipeline: one-hot encoding
cat_pipe = Pipeline([('encoder', OneHotEncoder(handle_unknown='ignore',sparse_output=False))])

# Pipeline: low variance and standardize
num_pipe = Pipeline([('standardize', StandardScaler())])

# combine all pipelines
all_pipe = make_pipeline(ColumnTransformer([('cat', cat_pipe, cat_vars),('num', num_pipe, num_vars)]))
X_train = all_pipe.fit_transform(X_train)
X_test = all_pipe.fit_transform(X_test)
X_val = all_pipe.fit_transform(X_val)
```
